```python
trans_utts = {}
dlg_utts = {}

def preproc(s):
    s = re.sub(r"<contraction .*>", "", s)
    return s
                                                                                                                                                                                                                                                                                                                                                    
for transfile in tqdm(list(glob.glob(os.path.join(DATA_PATH, "LDC2002T43", "reference", "english", "*.txt"))), desc="Load transcript"):
    dlgid = os.path.basename(transfile)[3:7]                                                                                                                       
    if dlgid not in trans_utts: trans_utts[dlgid] = []
    if dlgid not in dlg_utts: dlg_utts[dlgid] = []
        
    for line in open(transfile).read().split('\n'):
        if line.strip() == "": continue
        if line[0] == "#": continue
        line = line.lower().split(' ', 3)
        dlg_utts[dlgid].append(dict(
            start=float(line[0]),
            end=float(line[1]),
            caller=line[2][0],
            trans_words=line[3].split(' '),
        ))
        print(preproc(line[3]))
```

```python
# Example of a conversation
utts = [[utt['caller'], utt['start'], utt['end'], ' '.join(utt['trans_words'])] for utt in dlg_utts[random.choice(list(dlg_utts.keys()))]]
pd.set_option('display.max_rows', 10)
pd.DataFrame(utts, columns=["caller", "start", "end", "transcript"])
```
