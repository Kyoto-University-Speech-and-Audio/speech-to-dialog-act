```python
with open("/n/sd7/trung/csp/da_predicted_text_5_256.csv") as f:
    res2 = f.read().split('\n')[:-1]
    res2 = [line.split('\t') for line in res2]
print(len(res2))
```

```python
with open('/n/sd7/trung/csp/data/swbd/vocab/words_swda_full_old.txt') as f:
    vocab = f.read().split('\n')
```

```python
def decode(s):
    return ' '.join([vocab[int(_s)] for _s in s.split(' ')])
da_tags = list({'fo_o_fw_"_by_bc': 0, 'qw': 1, 'h': 2, 'sd': 3, 'sv': 4, 'b': 5, 'x': 6, '%': 7, '+': 8, 'qy': 9, 'qrr': 10, 'na': 11, 'bk': 12, 'ba': 13, 'ny': 14, '^q': 15, 'aa': 16, 'nn': 17, 'fc': 18, 'ad': 19, 'qo': 20, 'qh': 21, 'no': 22, 'ng': 23, '^2': 24, 'bh': 25, 'qy^d': 26, 'br': 27, 'b^m': 28, '^h': 29, 'bf': 30, 'fa': 31, 'oo_co_cc': 32, 'ar': 33, 'bd': 34, 't1': 35, 'arp_nd': 36, 't3': 37, 'ft': 38, '^g': 39, 'qw^d': 40, 'fp': 41, 'aap_am': 42}.keys())
```

```python
print(len(res1))
tab = []
c1, c2, c3 = 0, 0, 0
for r1, r2 in zip(res1, res2):  
    #if True:
    if r2[-2] != r1[-2] and r1[-2] == r1[2] and decode(r1[1]) != decode(r1[0]):
        logit1 = np.array([float(x) for x in r1[-1].split(' ')])
        prob1 = np.exp(logit1) / sum(np.exp(logit1))
        logit2 = np.array([float(x) for x in r2[-1].split(' ')])
        prob2 = np.exp(logit2) / sum(np.exp(logit2))
        
        tab.append((
            decode(r1[0]), 
            decode(r1[1]), 
            da_tags[int(r1[2])], 
            "%s (%.1f)" % (da_tags[int(r1[-2])], np.max(prob1) * 100), 
            "✔" if r1[-2] == r1[-3] else "✖",
            "%s (%.1f)" % (da_tags[int(r2[-2])], np.max(prob2) * 100),
            "✔" if r2[-2] == r1[-3] else "✖",
            da_tags[np.argmax(prob1 + prob2)],
        ))
        
        if r1[-2] == r1[-3]: c1 += 1
        if r2[-2] == r2[-3]: c2 += 1
        if np.argmax(prob1 + prob2) == int(r1[-3]): c3 += 1
    #if r1[-1] == r1[-2]: cr1 += 1
    #if r2[-1] == r2[-2]: cr2 += 1
#print(cr1, cr2)
pd.set_option('display.max_rows', 5000)
print("num of rows", len(tab))
print("da1 accuracy %.2f" % (c1 / len(res1) * 100))
print("da2 accuracy %.2f" % (c2 / len(res1) * 100))
print("vote %.2f" % (c3 / len(res1) * 100))
pd.DataFrame(tab, columns=["ground-truth text", "predicted_text", "da", "r1_da", "", "r2_da", "", "vote"])
```
